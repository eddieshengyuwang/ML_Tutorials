# K-Means Clustering and PCA
<br>
In this tutorial, we will explore an unsupervised learning algorithm called **K-Means Clustering** and a data compression algorithm called **Principal Component Analysis (PCA)**.

We start off with K-Means clustering. Since it is an unsupervised learning algorithm, the dataset we use will not have any labels. The goal of clustering is to find a "clustered" (lack of better word) pattern to our unlabelled data points. Let's begin by visualizing our dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io
from random import shuffle

data1 = scipy.io.loadmat("ex7data2.mat")
X = data1["X"]
plt.scatter(X[:,0], X[:,1], facecolors='none', color='b')
plt.show()
```

Again notice that the data points do not have labels to them. Visually, we can make out three different clusters of data. The k-means clustering algorithm allows a computer find this out too. 

Two initialize the algorithm, we randomly choose *k* points on the graph called the **cluster centroids**. In this assignment, we assign *k = 3*, which means that we will be finding three clusters.

Next, we define k-means clustering as an iterative algorithm with two steps:

1. **Cluster assignment step**: Classify every data point as "belonging" to a cluster centroid. A data point belongs to a particular cluster centroid if it is closest to that cluster centroid, defined by the following distance equation (also known as the Euclidean distance squared):
$$
d(x,y) = (x_1 - y_1)^2 + (x_2 - y_2)^2 + ... + (x_n - y_n)^2
$$


2. **Move centroid step**: Find the average of all the clustered points and move the respective cluster centroid to that average position. For example if two points (2, 5) and (3, 7) are in a cluster, the new centroid for that cluster will be at point (1.5, 6).

And repeat the above two steps until convergence! We start with implementing the cluster assignment step.

```python
def findClosestCentroids(X, centroids):
    m = X.shape[0]
    c = np.zeros(m)
    for i in range(m):
        coord = X[i]
        distances = list(map(lambda cent: euclid_sq(cent, coord), centroids))
        c[i] = np.argmin(distances)
    return c

def euclid_sq(centroid, coord):
    # centroid and coord should be same dimension
    # computes euclidean distance squared between centroid and coord 
    n = len(centroid)
    sum = 0
    for i in range(n):
        sum += (coord[i] - centroid[i])**2
    return sum

K = 3
initial_centroids = np.array([[3,3],[6,2],[8,5]]) # randomly assignment initial centroids
idx = findClosestCentroids(X, initial_centroids)
print(idx[:3])
```

`idx` is a ${\rm I\!R}^m$ vector that labels each training example to be in a cluster. Since Python is zero indexed, our output matches the one on the assignment. Next we will implement the move centroid step.

```python
def computeCentroids(X, idx, K):
    m = X.shape[0]
    n = X.shape[1]
    centroids = np.zeros((K,n))
    for i in range(K):
        # get all indices in idx where elements are = i
        pts = X[idx == i]
        mu = np.mean(pts, axis=0)
        centroids[i] = mu
    return centroids

computeCentroids(X, idx, K)
```

So after the first iteration, our centroid positions changed from [[3, 3], [6, 2], [8, 5]] to [[2.42830111, 3.15792418], [5.81350331, 2.63365645], [7.11938687, 3.6166844]]. Finally we implement the whole k-means algorithm and show the movements of the centroids along the way.

```python
def runkMeans(X, initial_centroids, max_iters, plot_progress):
    m, n = X.shape[0], X.shape[1]
    K = len(initial_centroids)
    centroidsLst = [initial_centroids]
    prev_centroidsLst = [initial_centroids] 
    idx = np.zeros(m)
    
    for i in range(max_iters):
        idx = findClosestCentroids(X, centroidsLst[i])
        
        if plot_progress:
            plotProgresskMeans(X, centroidsLst, prev_centroidsLst, idx, K, i)
            prev_centroidsLst.append(centroidsLst[i])
        
        newCentroids = computeCentroids(X, idx, K)
        centroidsLst.append(newCentroids)
    
    if plot_progress:
        plotFinalkMeans(X, centroidsLst[-1], idx, K, max_iters)
    
    return centroidsLst[-1], idx

def plotFinalkMeans(X, centroids, idx, K, max_iters):
    # plotting final k-Means visualization without showing steps
    colors = ["r","y","c"] 
    # different color for each cluster,
    # hard coded to 3 colors for this assignment 
    
    for i in range(K):
        pts = X[idx == i]
        plt.scatter(pts[:,0], pts[:,1], color=colors[i], facecolors='none', s=25)

    plt.scatter(centroids[:,0], centroids[:,1], marker="x", color="k",s=50)

    plt.title("Final plot after " + str(max_iters) + " iterations")
    plt.show()
    return

def plotProgresskMeans(X, centroidsLst, prev_centroidsLst, idx, K, currIter):
    colors = ["r","y","c"] 
    
    for i in range(K):
        pts = X[idx == i]
        plt.scatter(pts[:,0], pts[:,1], color=colors[i], facecolors='none', s=25)
    
        numCentroids = len(centroidsLst)
        for j in range(numCentroids):
            plt.plot([prev_centroidsLst[j][i][0], centroidsLst[j][i][0]],
                     [prev_centroidsLst[j][i][1], centroidsLst[j][i][1]],
                     marker="x", color="k")
      
    plt.title("Iteration " + str(currIter + 1))
    plt.show()

max_iters = 10
finalCentroid, finalIdx = runkMeans(X, initial_centroids, max_iters, True)
```

And ta-da! We classified three clusters successfully. Let's test our algorithm again by randomly initializing our cluster centroids. We randomly pick K data points from X to do this.

```python
def kMeansInitCentroids(X, K):
    # randomly shuffles elements in X
    # then takes first K elements from random X
    
    np.random.shuffle(X.copy()) # make sure you make a copy!!
    centroids = X[:K]
    return centroids

randomCentroids = kMeansInitCentroids(X, K)
print(randomCentroids)

finalCentroid, finalIdx = runkMeans(X, randomCentroids, max_iters, True)
```

```python
# PCA section

data2 = scipy.io.loadmat('ex7data1.mat')
X2 = data2['X']

plt.scatter(X2[:,0], X2[:,1])
plt.show()
```

```python
def featureNormalize(X):
    X_norm = X - np.mean(X, axis=0)
    X_norm = X_norm / np.std(X_norm, axis=0)
    return X_norm

X2_norm = featureNormalize(X2)

def pca(X):
    m = X.shape[0]
    n = X.shape[1]
    cov_matrix = np.dot(X.T,X) / m
    u, s, vh = np.linalg.svd(cov_matrix, full_matrices=True)
    return u,s

U,S = pca(X2_norm)

print("Top eigenvector is: ", U[:,0])    
```

```python
def projectData(X, U, K):
    U_reduce = U[:,:K] # n x k
    projections = np.dot(X, U_reduce) # X is m x n
    return projections

K = 1
Z = projectData(X2_norm, U, K)
Z[0]
```

```python
def recoverData(Z, U, K):
    U_reduce = U[:,:K] # n x k
    Z = np.matrix(Z) # m x 1
    X_approx = np.dot(Z,U_reduce.T)
    return X_approx

X_rec = recoverData(Z, U, K)
X_rec[0]
```

```python
plt.scatter(X2_norm[:,0], X2_norm[:,1], facecolors="none", edgecolors='b')
plt.scatter(X_rec[:,0], X_rec[:,1], facecolors="none", edgecolors='r')
plt.axis([-3,3,-3,3])

#plotting lines
for i in range(X2_norm.shape[0]):
    plt.plot([X2_norm[i,0], X_rec[i,0]],[X2_norm[i,1],X_rec[i,1]], "k--")

plt.gca().set_aspect('equal') # square shape 
plt.show()
```

```python
# face image part

data3 = scipy.io.loadmat("ex7faces.mat")
X3 = data3["X"] 
print(X3.shape)
```

```python
def displayData(X):
    # figure will be (32*10) x (32*10)
    
    figure = np.zeros((320, 320))
    xIndex = 0
    yIndex = 0
    
    for row in X[:100]:
        if xIndex == 320:
            yIndex += 32
            xIndex = 0
        img = np.reshape(row, (32,32)).T
        figure[xIndex:xIndex+32, yIndex:yIndex+32] = img
        xIndex += 32
        
    plt.figure(figsize=(5,5))
    plt.imshow(figure, cmap='gray')
    plt.show()
    return

displayData(X3)
```
