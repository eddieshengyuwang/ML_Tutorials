In this last assignment of the course, we will look at anomaly detection and recommendation engine algorithms. This post will cover the anomaly detection portion of the assignment.

An **anomaly** is a data point which deviates significantly from the rest of the dataset. Anomaly detection is an important process for many machine learning applications, such as detecting banking fraud or manufacturing defects. The overall problem statement for anomaly detection is: given an unlabelled training set, for a *new* data point can we determine if it is an anomaly?

To start the section, let's visualize our dataset which measures latency vs throughput for different computer servers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io

data1 = scipy.io.loadmat("ex8data1.mat")
X1 = data1["X"]

def plotData(X):
    plt.scatter(X[:,0], X[:,1], color="b", marker="x",s=7)
    plt.axis([0, 30, 0, 30])
    plt.xlabel("Latency (ms)")
    plt.ylabel("Throughput (mb/s)")

plotData(X1)
plt.show()
```

From the above graph, many points are concentrated together with a few scattered exceptions. We will now build an algorithm that classifies those scattered points as anomalies.

To do so, we assume that each feature *j* (of dataset X) is normally distributed by mean $\mu_j$ and variance $\sigma^2_j$. That is we assume that features ($x_1, x_2, x_3, ...$) are:

$$x_1 \sim \mathcal{N}(\mu_1, \sigma^2_1) \\
x_2 \sim \mathcal{N}(\mu_2, \sigma^2_2) \\ 
x_3 \sim \mathcal{N}(\mu_3, \sigma^2_3) \\ 
...
$$

We find (or *estimate*) the values for $\mu_1, \mu_2, \mu_3, ...$ and $\sigma^2_1, \sigma^2_2, \sigma^2_3, ...$ by using the current values for those features $x_1, x_2, x_3, ...$ in our training set. 

That is, $\mu_1 = mean(x_1), \mu_2 = mean(x_2), \mu_1 = mean(x_2), ... $ and $\sigma^2_1 = variance(x_1), \sigma^2_2 = variance(x_2), \sigma^2_3 = variance(x_3), ...$. 

This process is formally known as finding the **maximum likelihood estimates**. The next function will find the MLEs for $\mu$ and $\sigma^2$ based on our training set, where 
$$
\mu = \begin{bmatrix}
           \mu_{1} \\
           \mu_{2} \\
           \vdots \\
           \mu_{n}
         \end{bmatrix}
~~~~~~~~~~~~~~~~~~~~~~~~
\sigma^2 = \begin{bmatrix}
           \sigma^2_{1} \\
           \sigma^2_{2} \\
           \vdots \\
           \sigma^2_{n}
         \end{bmatrix}
$$


```python
def estimateGaussian(X): # Gaussian is synonymous with Normal 
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    sigma2 = np.square(sigma)
    return mu, sigma2

mu, sigma2 = estimateGaussian(X1)

print("mu: ", mu, "\nsigma^2: ", sigma2)
# vectors of length 2 since we only have 2 features in this dataset
```

Next we define the **probability density function** *p* for a data point *x* $\in \rm I\!R^n$ as:

\begin{align}
p(x) &= p(x_1; \mu_1,\sigma^2_1)p(x_2; \mu_2,\sigma^2_2)...p(x_n; \mu_n,\sigma^2_n) \\
    &= \prod_{j=1}^n p(x_j; \mu_j, \sigma^2_j) \\
    &= \prod_{j=1}^n\frac{1}{\sqrt{2\pi}\sigma_j} \exp(-\frac{(x_j - u_j)^2}{2\sigma_j^2})
\end{align}

Note that *p* is **not** probability, *p* is probability density!

Let's now find the probability densities for all our training examples using our estimated $\mu$ and $\sigma^2$ found above.

```python
def multivariateGaussian(X, mu, sigma2):
    n = len(mu)
    if sigma2.ndim == 1:
        sigma2 = np.diag(sigma2)
    constant = (2*np.pi)**(-n/2) * np.linalg.det(sigma2)**(-0.5) 
    inside_exp = -0.5 * np.diag(np.dot(np.dot(X - mu, np.linalg.inv(sigma2)), (X - mu).T)) 
    return constant * np.exp(inside_exp)

p = multivariateGaussian(X1, mu, sigma2)
print(p)
print(p.shape)
```

So *p* above represents all the probability density values for all 307 training examples. Since our dataset has 2 features, we can visualize a 2D normal distribution (formally known as a **bivariate normal distribution**) from our probability densities. It will look something similar to this (except our "peak" location and "smoothness" will differ):

But instead of making that plot, we will draw contour lines for our training data.

```python
def visualizeFit(X, mu, sigma2):
    x1Range = np.linspace(0, 30, 100)
    x2Range = np.linspace(0, 30, 100)
    XX,YY = np.meshgrid(x1Range, x2Range)
    coordinates = np.c_[XX.ravel(), YY.ravel()]
    Z = multivariateGaussian(coordinates, mu, sigma2)
    Z = Z.reshape(XX.shape)
    plotData(X)
    cont_levels = [10**exp for exp in range(-20,0,3)]
    plt.contour(XX,YY,Z,cont_levels)
    
visualizeFit(X1, mu, sigma2)
plt.show()
```

In the above graph, points that are further away from the middle yellow circle correspond to points of lower probability density, whereas points inside the yellow circle correspond to high values of *p*.

Thus, we will define a constant **$\epsilon$** such that if $p(x) \lt \epsilon$, then *x* is an anomaly. The next step is to find out what this $\epsilon$ value is.

To do this, we will use a validation dataset provided by the assignment, which does have labels on whether a data point is an anomaly or not! (Usually the training set will not have labels, but a validation set will have a few labels for anomalies, since they are so rare anyways). We then select the best $\epsilon$ value that best classifies anomalies based on the highest **F1 score**.

```python
X1val = data1["Xval"]
Y1val = data1["yval"]
pval = multivariateGaussian(X1val, mu, sigma2) # probability densities for each example in validation set
```

The F1 Score is another metric (like accuracy) to evaluate how well a classifier performs. Let's take a look at scenario where accuracy may not be a good measurement of performance.

Suppose we have very *skewed* labelled dataset of 100 entries where all the labels are 0 except for three entries that are labelled 1. Then suppose we have a classification algorithm that achieves 97% accuracy on that dataset. At first glance, our model is great because it is predicting 97% of our dataset correctly! But what if the algorithm was just naively labelling each data point as 0? Since our dataset only has 3 positive examples, the classification accuracy of this algorithm is 97%, but it is obviously a bad model since it is naively classifying each data point as 0.

In these situations, the F1 Score is a better measure of performance, and is defined by:

$$
F1~ Score = 2 \Big(\frac{precision \times recall}{precision + recall}\Big)
$$

Where **precision** is defined as $\frac{\text{true positives}}{\text{true positives + false positives}}$ and **recall** is defined as $\frac{\text{true positives}}{\text{true positives + false negatives}}$.


- **True positives** are examples where we predicted 1 and in the dataset are labelled 1 as well.
- **True negatives** are examples where we predicted 0 and in the dataset are labelled 0 as well.
- **False positives** are examples where we predicted 1 and in the dataset are labelled 0 (thus classified incorrectly).
- **False negatives** are examples where we predicted 0 and in the dataset are labelled 1 (thus classified incorrectly).

Note that by convention, the less frequent class is labelled 1, so in this case anomalies are deemed 1 in the given validation set.

In general, *precision* is "out of all examples where we predicted 1 (true positives + false positives), what fraction actually is 1 (true positive)" whereas *recall* is "out of all the examples that are 1 (true positives + false negatives), what fraction did we predict 1 correctly (true positives)".

Lastly, higher F1 Score implies better performance of our algorithm. So in the following function, we will make predictions with different $\epsilon$ values and select the $\epsilon$ that gives the highest F1 Score.

```python
def selectThreshold(yval, pval):
    bestEpsilon = 0
    bestF1 = 0
    F1 = 0
    stepsize = (np.max(pval) - np.min(pval)) / 1000
    for epsilon in np.arange(np.min(pval), np.max(pval), stepsize):
        predictions = (pval < epsilon).astype(int).reshape((-1,1))
        
        tp = np.logical_and(predictions == 1, yval == 1).astype(int)
        tn = np.logical_and(predictions == 0, yval == 0).astype(int) # note we never use this
        fp = np.logical_and(predictions == 1, yval == 0).astype(int)
        fn = np.logical_and(predictions == 0, yval == 1).astype(int)
        precision = np.sum(tp) / (np.sum(tp) + np.sum(fp))
        recall = np.sum(tp) / (np.sum(tp) + np.sum(fn))
        
        F1 = 2*precision*recall/(precision + recall)
        if F1 > bestF1:
            bestF1 = F1
            bestEpsilon = epsilon
    return bestEpsilon, bestF1

epsilon, f1 = selectThreshold(Y1val, pval)
print("best epsilon: ", epsilon,"\nbest F1: ", f1)
```

All the numbers align with the ones stated in the assignment. Next we will circle the anomalies in the training set based on the epsilon value we found from the validation set.

```python
outliers = np.argwhere(p < epsilon).ravel()
print("outlier indices: ", outliers)
```

```python
X1[outliers] # data points that are anomalies
```

```python
visualizeFit(X1, mu, sigma2)
plt.scatter(X1[outliers,0], X1[outliers,1], marker="o",
            facecolor="none", color="r", s=150)
plt.show()
```

Cool! In the final section of this post, we will examine a more realistic dataset with more than 2 features and detect anomalies from there.

```python
data2 = scipy.io.loadmat("ex8data2.mat")
X2 = data2["X"]
Xval2 = data2["Xval"]
yval2 = data2["yval"]

mu2, sigma2_2 = estimateGaussian(X2)
p2 = multivariateGaussian(X2, mu2, sigma2_2)
pval2 = multivariateGaussian(Xval2, mu2, sigma2_2)

epsilon2, f1_2 = selectThreshold(yval2, pval2)

print("X shape: ", X2.shape)
print("best epsilon: ", epsilon2)
print("best F1: ", f1_2)
print("outliers found: ", np.sum(p2 < epsilon2))
```
